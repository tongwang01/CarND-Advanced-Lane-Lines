# Advanced Lane Finding Project


Output
* Output video: lane lines detected and drawn, numerical estimation of lane curvature and vehicle position displayed.
* Output test images
* Intermediate images
* Writeup

Tasks
* Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
* Apply a distortion correction to raw images.
* Use color transforms, gradients, etc., to create a thresholded binary image.
* Apply a perspective transform to rectify binary image ("birds-eye view").
* Detect lane pixels and fit to find the lane boundary.
* Determine the curvature of the lane and vehicle position with respect to center.
* Warp the detected lane boundaries back onto the original image.
* Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.

## Compute the camera calibration matrix and distortion coefficients given a set of chessboard images, and apply distortion correction to raw images

```python
def calibrate_camera(path, corners_shape = (9, 6)):
    """Calibrate camera based on a number of images of chessboards.
    
    Args:
        path: path of calibration images
        corners_shape: shape of corners
    
    Returns:
        mtx: camera matrix
        dist: distortion coefficients
    """
    # Set object points
    objp = np.zeros((6*9,3), np.float32)
    objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)
    
    # Arrays to store image points and object points
    objpoints = []
    imgpoints = []
    
    # Read calibration image files, find corners, append to 
    images = glob.glob(path)
    for fname in images:
        img = mpimg.imread(fname)
        gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
        
        # Find corners
        ret, corners = cv2.findChessboardCorners(gray, (9,6),None)
        
        # If found, add object points, image points
        if ret == True:
            objpoints.append(objp)
            imgpoints.append(corners)
    
    # Calibrate camera
    ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1], None, None)
    
    return mtx, dist


def undistort(img, mtx, dist):
    """Undistrot an image given camera matrix and distortion coefficients.
    
    Returns:
        undistorted: undistorted image
    """
    return cv2.undistort(img, mtx, dist, None, mtx)
```

```python
# Calibrate camera
mtx, dist = calibrate_camera(path = "./camera_cal/calibration*.jpg")

# Undistort calibration images, write output
try:
    os.mkdir("./camera_cal_output")
except:
    pass

calibration_images = glob.glob("./camera_cal/calibration*.jpg")
for f in calibration_images:
    img = mpimg.imread(f)
    undist = undistort(img, mtx, dist)
    mpimg.imsave(f.replace("/camera_cal/", "/camera_cal_output/"), undist)
```

## Use color transforms, gradients, etc., to create a thresholded binary image

### Try a combined gradient & color thresholds

Results on test images seem better with this approach (applying combined gradient and color thresholds) than applying each individual threshold or applying the two sequentially, so we will use this one.

Parameters tuned manually by looking at test images.

```python
def gradient_and_color_threshold(img, 
                                 s_threshold=(20, 255),
                                 sobel_kernel=3,
                                 mag_threshold=(20, 255),
                                 dir_threshold=(0.5, 1.5)):
    """Apply color threshold on the S channel, magnitulte and direction thresholds on the gradient.
    
    Returns:
        binary_output: a thresholded binary output image
    """
    # Color
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    S = hls[:,:,2]
    s_mask = (S > s_threshold[0]) & (S <= s_threshold[1])

    # Gradient
    #gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    gray = S
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1)
    sobel_mag = np.sqrt(sobelx**2 + sobely**2)
    scaled_sobel = np.uint8(255*sobel_mag/np.max(sobel_mag))
    grad_direction = np.arctan2(np.absolute(sobelx), np.absolute(sobely)) 
    
    mag_mask = (scaled_sobel >= mag_threshold[0]) & (scaled_sobel <= mag_threshold[1])
    dir_mask = (grad_direction > dir_threshold[0]) & (grad_direction < dir_threshold[1])
   
    # Binary output    
    binary_output = np.zeros_like(S)
    binary_output[s_mask & mag_mask & dir_mask] = 1
    
    return binary_output
```

```python
# Test it on one test image
f = "test3.jpg"
original = mpimg.imread(os.path.join("./test_images/", f))
cur = undistort(original, mtx, dist)
cur = gradient_and_color_threshold(cur,
                                   s_threshold=(10, 255),
                                   sobel_kernel=3,
                                   mag_threshold=(10, 255),
                                   dir_threshold=(0.2, 1.5))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 16))
ax1.imshow(original)
ax1.set_title("Original image \n" + f)
ax2.imshow(cur, cmap="gray")
ax2.set_title("Thresholded image \n" + f)
fig.show()
```

## Apply a perspective transform to rectify binary image ("birds-eye view")

```python
# Calibrate perspective transform on a straight line test image

# Read in image
f = "straight_lines2.jpg"
original = mpimg.imread(os.path.join("./test_images_output_undistort/", f))

%matplotlib
plt.imshow(original, cmap="gray")

# Identify source points by eyeballing a straight line test image
src = np.float32([
    [298, 660],
    [555, 478],
    [732, 478],
    [1015, 660]])

for point in src:
    plt.plot(point[0], point[1], ".", markersize=15, color="red")
    
    
# Identify destination points
dst = np.float32([
    [298, 700],
    [298, 10],
    [1015, 10],
    [1015, 700]])

for point in dst:
    plt.plot(point[0], point[1], ".", markersize=15, color="blue")
    
# Compute perspective matrix
M = cv2.getPerspectiveTransform(src, dst)
Minv = cv2.getPerspectiveTransform(dst, src)
```

```python
def perspective_transform(img, M=M):
    img_size = (img.shape[1], img.shape[0])
    return cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
```

## Detect lane pixels and fit to find the lane boundary

```python
def find_lane_pixels(binary_warped):
    # Take a histogram of the bottom half of the image
    histogram = np.sum(binary_warped[binary_warped.shape[0]//2:,:], axis=0)
    # Create an output image to draw on and visualize the result
    out_img = np.dstack((binary_warped, binary_warped, binary_warped))
    # Find the peak of the left and right halves of the histogram
    # These will be the starting point for the left and right lines
    midpoint = np.int(histogram.shape[0]//2)
    leftx_base = np.argmax(histogram[:midpoint])
    rightx_base = np.argmax(histogram[midpoint:]) + midpoint

    # HYPERPARAMETERS
    # Choose the number of sliding windows
    nwindows = 9
    # Set the width of the windows +/- margin
    margin = 100
    # Set minimum number of pixels found to recenter window
    minpix = 50

    # Set height of windows - based on nwindows above and image shape
    window_height = np.int(binary_warped.shape[0]//nwindows)
    # Identify the x and y positions of all nonzero pixels in the image
    nonzero = binary_warped.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    # Current positions to be updated later for each window in nwindows
    leftx_current = leftx_base
    rightx_current = rightx_base

    # Create empty lists to receive left and right lane pixel indices
    left_lane_inds = []
    right_lane_inds = []

    # Step through the windows one by one
    for window in range(nwindows):
        # Identify window boundaries in x and y (and right and left)
        win_y_low = binary_warped.shape[0] - (window+1)*window_height
        win_y_high = binary_warped.shape[0] - window*window_height
        win_xleft_low = leftx_current - margin
        win_xleft_high = leftx_current + margin
        win_xright_low = rightx_current - margin
        win_xright_high = rightx_current + margin
        
        """
        # Draw the windows on the visualization image
        cv2.rectangle(out_img,(win_xleft_low,win_y_low),
        (win_xleft_high,win_y_high),(0,255,0), 2) 
        cv2.rectangle(out_img,(win_xright_low,win_y_low),
        (win_xright_high,win_y_high),(0,255,0), 2)
        """ 
        
        # Identify the nonzero pixels in x and y within the window #
        good_left_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & 
        (nonzerox >= win_xleft_low) &  (nonzerox < win_xleft_high)).nonzero()[0]
        good_right_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & 
        (nonzerox >= win_xright_low) &  (nonzerox < win_xright_high)).nonzero()[0]
        
        # Append these indices to the lists
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
        
        # If you found > minpix pixels, recenter next window on their mean position
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds) > minpix:        
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))

    # Concatenate the arrays of indices (previously was a list of lists of pixels)
    try:
        left_lane_inds = np.concatenate(left_lane_inds)
        right_lane_inds = np.concatenate(right_lane_inds)
    except ValueError:
        # Avoids an error if the above is not implemented fully
        pass

    # Extract left and right line pixel positions
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds]

    return leftx, lefty, rightx, righty, out_img


def fit_polynomial(binary_warped):
    # Find our lane pixels first
    leftx, lefty, rightx, righty, out_img = find_lane_pixels(binary_warped)

    # Fit a second order polynomial to each using `np.polyfit`
    try:
        left_fit = np.polyfit(lefty, leftx, 2)
        right_fit = np.polyfit(righty, rightx, 2)
    except:
        left_fit = np.array([0, 0, 0])
        right_fit = np.array([0, 0, 0])

    # Generate x and y values for plotting
    ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0] )
    try:
        left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
        right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
    except TypeError:
        # Avoids an error if `left` and `right_fit` are still none or incorrect
        print('The function failed to fit a line!')
        left_fitx = 1*ploty**2 + 1*ploty
        right_fitx = 1*ploty**2 + 1*ploty

    ## Visualization ##
    # Colors in the left and right lane regions
    out_img[lefty, leftx] = [255, 0, 0]
    out_img[righty, rightx] = [0, 0, 255]

    # Plots the left and right polynomials on the lane lines
    #plt.plot(left_fitx, ploty, color='yellow')
    #plt.plot(right_fitx, ploty, color='yellow')

    return out_img, ploty, left_fitx, right_fitx
```

```python
# Test it on one test image
f = "test2.jpg"
original = mpimg.imread(os.path.join("./test_images/", f))
cur = undistort(original, mtx, dist)
cur = gradient_and_color_threshold(cur)
cur = perspective_transform(cur)
cur, ploty, left_fitx, right_fitx = fit_polynomial(cur)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 16))
ax1.imshow(original)
ax1.set_title("Original image \n" + f)
ax2.imshow(cur, cmap="gray")
ax2.plot(left_fitx, ploty, color='yellow')
ax2.plot(right_fitx, ploty, color='yellow')
ax2.set_title("Detected lines image \n" + f)
fig.show()
```

## Determine the curvature of the lane and vehicle position with respect to center

```python
def measure_curvature_real(ploty, left_fitx, right_fitx):
    '''
    Calculates the curvature of polynomial functions in meters.
    '''
    # Define conversions in x and y from pixels space to meters
    ym_per_pix = 3/150 # meters per pixel in y dimension
    xm_per_pix = 3.7/700 # meters per pixel in x dimension
        
    left_fit_cr = np.polyfit(ploty*ym_per_pix, left_fitx*xm_per_pix, 2)
    right_fit_cr = np.polyfit(ploty*ym_per_pix, right_fitx*xm_per_pix, 2)
    
    # Define y-value where we want radius of curvature
    # We'll choose the maximum y-value, corresponding to the bottom of the image
    y_eval = np.max(ploty)
    
    # Calculation of R_curve (radius of curvature)
    left_curverad = ((1 + (2*left_fit_cr[0]*y_eval*ym_per_pix + left_fit_cr[1])**2)**1.5) / np.absolute(2*left_fit_cr[0])
    right_curverad = ((1 + (2*right_fit_cr[0]*y_eval*ym_per_pix + right_fit_cr[1])**2)**1.5) / np.absolute(2*right_fit_cr[0])
    
    return left_curverad, right_curverad
```

```python
f = "test2.jpg"
original = mpimg.imread(os.path.join("./test_images/", f))
undist = undistort(original, mtx, dist)
cur = gradient_and_color_threshold(undist)
cur = perspective_transform(cur)
_, ploty, left_fitx, right_fitx = fit_polynomial(cur)

left_curverad, right_curverad = measure_curvature_real(ploty, left_fitx, right_fitx)

print(round(left_curverad/1000,1), 'km', round(right_curverad/1000,1), 'km')
```

## Warp the detected lane boundaries back onto the original image

```python
f = "test5.jpg"
original = mpimg.imread(os.path.join("./test_images/", f))
undist = undistort(original, mtx, dist)
cur = gradient_and_color_threshold(undist)
cur = perspective_transform(cur)
_, ploty, left_fitx, right_fitx = fit_polynomial(cur)

# Create an image to draw the lines on
warp_zero = np.zeros_like(cur).astype(np.uint8)
color_warp = np.dstack((warp_zero, warp_zero, warp_zero))

# Recast the x and y points into usable format for cv2.fillPoly()
pts_left = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
pts = np.hstack((pts_left, pts_right))

# Draw the lane onto the warped blank image
cv2.fillPoly(color_warp, np.int_([pts]), (0,255, 0))

# Warp the blank back to original image space using inverse perspective matrix (Minv)
newwarp = cv2.warpPerspective(color_warp, Minv, (original.shape[1], original.shape[0])) 
# Combine the result with the original image
result = cv2.addWeighted(undist, 1, newwarp, 0.3, 0)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 16))
ax1.imshow(original)
ax1.set_title("Original image \n" + f)
ax2.imshow(result)
ax2.set_title("Detected lines image \n" + f)
fig.show()
```

## Put it all together

```python
def detect_lane_in_image(img):
    """Recognizes lanes in an image, and draw onto the image.
    
    Returns:
        output_image
    """
    # Undistort
    undist = undistort(img, mtx, dist)
    # Apply gradient and color thresholds
    cur = gradient_and_color_threshold(undist)
    # Perspective transform
    cur = perspective_transform(cur)
    # Fit polylines
    _, ploty, left_fitx, right_fitx = fit_polynomial(cur)
    
    # Draw lanes onto the original image
    warp_zero = np.zeros_like(cur).astype(np.uint8)
    color_warp = np.dstack((warp_zero, warp_zero, warp_zero))
    
    pts_left = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
    pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
    pts = np.hstack((pts_left, pts_right))
    
    cv2.fillPoly(color_warp, np.int_([pts]), (0,255, 0))

    newwarp = cv2.warpPerspective(color_warp, Minv, (original.shape[1], original.shape[0])) 
    output_image = cv2.addWeighted(undist, 1, newwarp, 0.3, 0)
    
    # Write curvature info
    left_curverad, right_curverad = measure_curvature_real(ploty, left_fitx, right_fitx)
    font = cv2.FONT_HERSHEY_SIMPLEX
    fontScale = 2
    fontColor = (255, 255, 255)
    cv2.putText(output_image, 'Left curvature: {:.2f} km'.format(left_curverad/1000), 
                (50, 50), font, fontScale, fontColor, 2)
    cv2.putText(output_image, 'Right curvature: {:.2f} km'.format(right_curverad/1000), 
                (50, 120), font, fontScale, fontColor, 2)

    return output_image
```

```python
# Test on one test image
f = "test2.jpg"
original = mpimg.imread(os.path.join("./test_images/", f))
cur = detect_lane_in_image(original)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 16))
ax1.imshow(original)
ax1.set_title("Original image \n" + f)
ax2.imshow(cur)
ax2.set_title("Detected lines image \n" + f)
fig.show()
```
